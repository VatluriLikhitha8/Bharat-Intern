```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score ,classification_report
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
```

```python
null_percentages = (df.isnull().sum()/len(df))*100
null_percentages
```

```python
max_null_colmn = null_percentages.idxmax()
max_null_colmn
```

```python
df.drop(columns=[max_null_colmn],inplace=True)
```

```python
df
```

```python
if df['Age'].dtype =='float64':
    df['Age'].fillna(df['Age'].mean(),inplace = True)
else :
    df['Age'].fillna(df['Age'].mode()[0],inplace = True)
```

```python
df['Sex'].replace(['male','female'],[0,1],inplace = True)
```

```python
df
```

```python
x = df[['Pclass','Age','Sex','Fare']]
x
```

```python
y = df['Survived']
y
```

```python
x_train,x_test,y_train,y_test = train_test_split (x, y, test_size = 0.25,random_state =32)
```

```python
model = LogisticRegression()
model.fit( x_train, y_train)
```

```python
y_prediction =model.predict( x_test) ;y_prediction
```

```python
ac = accuracy_score (y_test, y_prediction)
r = classification_report( y_test, y_prediction)
```

```python
print("Accuracy = {:.2f}%".format(ac*100))
```
